### Merges MMSE_Category by PTID to the dataset with the confirmed NP Patients.

Shows PTID, MMSE_Category, and NPADNC

As well as the 3 values that make up the NPADNC Composite value, NPNEUR, NPBRAAK, and NPTHAL.

```python
%sql
SELECT
  n.PTID,
  f.MMSE_Category,
  n.NPISCORE,
  n.NPADNC,
  n.NPNEUR,
  n.NPBRAAK,
  n.NPTHAL
FROM MERGED_NEUROPATH_Multimodal_ADNI n
LEFT JOIN final_Multimodal_ADNI f
  ON n.PTID = f.PTID;
```

Separate code that actually creates a table rather than viewing it like the code above

```python
%sql
CREATE OR REPLACE TABLE NP_with_MMSE_AND_NPISCORE AS
SELECT
  n.PTID,
  f.MMSE_Category,
  n.NPISCORE,
  n.NPADNC,
  n.NPNEUR,
  n.NPBRAAK,
  n.NPTHAL
FROM MERGED_NEUROPATH_Multimodal_ADNI n
LEFT JOIN final_Multimodal_ADNI f
  ON n.PTID = f.PTID;
```

```python
display(
    spark.sql(
        "DESCRIBE NP_with_MMSE_AND_NPISCORE"
    )
)
```

```python
%sql
SELECT COUNT(*) AS AD_Catg
FROM NP_with_MMSE_AND_NPISCORE
WHERE MMSE_Category IN ("Mild AD", "Moderate AD")

UNION

SELECT COUNT(*) AS Normal_Catg
FROM NP_with_MMSE_AND_NPISCORE
WHERE MMSE_Category IN ("Normal cognition / Possible MCI (Borderline)","Normal cognition (Fully intact)");
```

```python
%sql
SELECT
    (SELECT COUNT(*)
    FROM NP_with_MMSE_AND_NPISCORE
    WHERE MMSE_Category IN ("Mild AD", "Moderate AD")) AS AD_Catg,
    (SELECT COUNT(*)
    FROM NP_with_MMSE_AND_NPISCORE
    WHERE MMSE_Category IN ("Normal cognition / Possible MCI (Borderline)","Normal cognition (Fully intact)")) AS Normal_Catg;
```

```python
%sql
SELECT
    (SELECT COUNT(*)
    FROM NP_with_MMSE_AND_NPISCORE
    WHERE MMSE_Category IN ("Moderate AD")) AS Moderate_AD_Catg,
    (SELECT COUNT(*)
    FROM NP_with_MMSE_AND_NPISCORE
    WHERE MMSE_Category IN ("Mild AD")) AS Mild_AD_Catg,
    (SELECT COUNT(*)
    FROM NP_with_MMSE_AND_NPISCORE
    WHERE MMSE_Category IN ("Normal cognition / Possible MCI (Borderline)")) AS Borderline_Catg,
    (SELECT COUNT(*)
    FROM NP_with_MMSE_AND_NPISCORE
    WHERE MMSE_Category IN ("Normal cognition (Fully intact)")) AS Normal_Catg;
```
